```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
G = nx.DiGraph()

G.add_edges_from([('fred:there_1', 'fred:Cat')], label='rdf:type')
G.add_edges_from([('fred:there_1', 'fred:mat_1')], label='fred:on')
G.add_edges_from([('fred:there_1', 'quant:a')], label='quant:hasDeterminer')
G.add_edges_from([('fred:there_1', 'fred:there')], label='rdf:type')
```

```python
val_map = {'A': 1.0,
            'D': 0.5714285714285714,
            'H': 0.0}

values = [val_map.get(node, 0.45) for node in G.nodes()]
```

```python
edge_labels=dict([((u,v,), d['label'])
                 for u,v,d in G.edges(data=True)])
```

```python
pos=nx.spring_layout(G)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
nx.draw(G,pos, node_color = 'blue', node_size=2000,edge_cmap=plt.cm.Reds)
pylab.show()
```
