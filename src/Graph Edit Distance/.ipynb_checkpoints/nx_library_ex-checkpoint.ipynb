```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
G = nx.DiGraph()

G.add_edges_from([('A1', 'B1')], label='rdf:type', weight=1)
```

```python
G.add_edges_from([('A', 'B'),('C','D'),('G','D')], label='1')
G.add_edges_from([('B','C'),('E','F')], label='3')
G.add_edges_from([('C','F')], label='4')
```

```python
val_map = {'A': 1.0,
            'D': 0.5714285714285714,
            'H': 0.0}

values = [val_map.get(node, 0.45) for node in G.nodes()]
```

```python
edge_labels=dict([((u,v,), d['label'])
                 for u,v,d in G.edges(data=True)])
```

```python
pos=nx.spring_layout(G)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
nx.draw(G,pos, node_color = 'blue', node_size=2000,edge_cmap=plt.cm.Reds)
pylab.show()
```
